```python
import pandas as pd
data = pd.read_csv("/content/fifa21 raw data v2.csv") #Insert Data
data
```

The aim is to clean the data so that it is prepared for analysis and data visualisation whilst in third normal form (3NF).

```python
# Select relevant columns.
columns_to_select = ['Name', 'Age','Positions', 'PAC', 'PAS', 'PHY']
new_data = data[columns_to_select]
new_data.head(6)
```

```python
# Gather info on dataset.
new_data.info()
```

```python
# Describe Data:
new_data.describe()
```

```python
# Simplify columns.
new_data = new_data.rename(columns = {"Positions": "POS"})
```

```python
# Show Simplified Result:
new_data.head(6)
```

```python
# I want to analyse pace and physicality in a player, GK's woould be useless for this level of analysis so they will be removed.
new_data = new_data[~new_data['POS'].str.contains('GK', na=False)]
```

```python
new_data.shape
```

```python
new_data["POS"] = new_data["POS"].replace(r'\b(ST|RW|LW|CF|CAM)', 'Attacker', regex=True)
new_data["POS"] = new_data["POS"].replace(r'\b(CB|LB|LWB|RWB|CDM)', 'Defender', regex=True) # Simplify positions, simpler to analyse the data.
new_data["POS"] = new_data["POS"].replace(r'\b(CM|LM|RM)', 'Midfielder', regex=True)
# Used ChatGPT to convert "Attacker * 3" to just "Attacker"
new_data["POS"] = new_data["POS"].apply(lambda x: 'Attacker' if 'Attacker' in x else
                                         ('Defender' if 'Defender' in x else
                                          ('Midfielder' if 'Midfielder' in x else x)))
new_data.head(6)
```

```python
# Visualisation of data to show the distribution of age via Histogram:
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
plt.figure(figsize = (6, 4))
sns.histplot(new_data["Age"], bins = 20, kde = True)
plt.title("Histogram Distribution of Age: ")
plt.xlabel("Age: ")
plt.ylabel("Frequency: ")
plt.show()
```

```python
# Create bar plot using seaborn to display visualisation on player's pace and physicality.
while True:
  n1 = random.randint(0, len(new_data) - 1) # Gather random player 1.
  n2 = random.randint(0, len(new_data) - 1) # Gather random player 2.
  if n1 != n2 and new_data["POS"].iloc[n1] == new_data["POS"].iloc[n2]: # Ensure both randoms are not equal (that will create 2 of the same bar charts of the same player).
    break # Also ensure both players have the same position. Cannot compare a defender to an attacker as easily.
plt.figure(figsize = (4, 4))
sns.barplot(x = ["PHY", "PAC"], y = [new_data["PHY"].iloc[n1], new_data["PAC"].iloc[n1]])
plt.xlabel(" ")
plt.ylabel("Frequency/Score: ")
plt.title(f"Comparison of: {new_data['Name'].iloc[n1]}")
print("Name: ", new_data["Name"].iloc[n1], " | ", "Age: ", new_data["Age"].iloc[n1], " | ", "Pace: ", new_data["PAC"].iloc[n1], " | ", "Physicality: ", new_data["PHY"].iloc[n1], " | ", "Position: ", new_data["POS"].iloc[n1])
# Indicate other factors of data (position, name and age).
plt.show()
plt.figure(figsize = (4, 4))
sns.barplot(x = ["PHY", "PAC"], y = [new_data["PHY"].iloc[n2], new_data["PAC"].iloc[n2]])
plt.xlabel(" ")
plt.ylabel("Frequency/Score: ")
plt.title(f"Comparison of: {new_data['Name'].iloc[n2]}")
print("Name: ", new_data["Name"].iloc[n2], " | ", "Age: ", new_data["Age"].iloc[n2], " | ", "Pace: ", new_data["PAC"].iloc[n2], " | ", "Physicality: ", new_data["PHY"].iloc[n2], " | ", "Position: ", new_data["POS"].iloc[n2])
plt.show()
```

```python
data = data.rename(columns = {"↓OVA": "OVA"}) # Renaming.
new_data["OVA"] = data["OVA"] # Adding existing columns from the original dataset into the simplified version.
new_data["POT"] = data["POT"]
new_data
```

```python
plt.figure(figsize=(4, 4))
sns.lineplot(x = new_data["POT"], y=new_data["Age"], color = "red")
plt.title("Correlation between potential and age: ")
plt.show()
plt.figure(figsize=(4, 4))
sns.lineplot(x = new_data["OVA"], y=new_data["Age"], color = "green")
plt.title("Correlation between Overall Rating and age: ")
plt.show() # These correlations suggest that age has an impact on potential as well as overall rating.
```

```python
import plotly.express as px
new_data["Best Position"] = data["Best Position"]
bestplayers = new_data.nlargest(30, "OVA") # Best players by position
figure = px.bar(bestplayers, x = "Best Position", y = "OVA", color = "Age Group", hover_name = "Name", text = "Name", title = "Best players based off Position: ")
figure.show()
bestplayers = new_data.nsmallest(30, "OVA") # Worst players by position.
figure = px.bar(bestplayers, x = "Best Position", y = "OVA", color = "Age Group", hover_name = "Name", text = "Name", title = "Worst players based off Position: ")
figure.show()
# Colour coded by age group to indicate the younger the player, the more likely to have a lower rating.
#Missing Plots, posted on files.
```

```python
plt.figure(figsize =(6, 6))
bins = [0, 20, 24, 30, 36, 40]
labels = ["under 20", "20-24", "25-30", "31-35", "36+"] # Create age group labels to get an idea of the % of specific ages within the group.
new_data["Age Group"] = pd.cut(new_data["Age"], bins=bins, labels=labels, right=False)
age_group_counts = new_data["Age Group"].value_counts()
plt.pie(age_group_counts, labels= age_group_counts.index, autopct="%1.1f%%")
plt.title("% of Age Groups: ")
plt.show()
```
